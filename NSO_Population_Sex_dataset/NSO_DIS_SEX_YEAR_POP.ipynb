```python
import pandas as pd
```

```python
# Set display options to show all columns and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Reset display options to default values if needed
# pd.reset_option('display.max_columns')
# pd.reset_option('display.max_rows')
```

```python
# Assuming the CSV file is in the same folder as your Jupyter Notebook
file_name = 'NSO_DF_TOT_POP_BY_REG_DIST_LOC_1.5.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_name)
print(df)
```

```python
# Dropping extra columns
columns_to_drop = ['DATAFLOW', 'CSE_FREQ: Frequency', 'CSE_OBS_STATUS: Observation Status', 'CSE_OBS_NOTE: Observation Note']
df = df.drop(columns=columns_to_drop)
print(df)
```

```python
# Rename specified columns
column_mapping = {
    'CSE_REG_DIST_LOC: Region District Locality': 'District',
    'CSE_SEX: Sex': 'Sex',
    'TIME_PERIOD: Time Period': 'Year',
    'OBS_VALUE': 'Population'
}
df = df.rename(columns=column_mapping)
print(df)
```

```python
# Clean the 'District', 'Sex', and 'Year' columns
df['District'] = df['District'].str.extract(r': (\d+) -')
df['Sex'] = df['Sex'].str.extract(r'(\w)')
df['Year'] = df['Year'].str.extract(r'(\d+)-')
print(df)
```

```python
# Filter to records pertaining to districts only.
df['District'] = df['District'].astype(str)
df = df[df['District'].str.len() == 3]

# Keeping only the last digit
df['District'] = df['District'].str[-1]

print(df)
```

```python
# Ensuring 'Year' column is a numeric type
df['Year'] = df['Year'].astype(int)
print(df)
```

```python
# Drop records related to total in the 'Sex' column
df = df[df['Sex'] != 'T']

# Replace 'M' with 0 and 'F' with 1 in the 'Sex' column
df['Sex'] = df['Sex'].replace({'M': 0, 'F': 1})
df['Sex'] = pd.to_numeric(df['Sex'])

print(df)
```

```python
# Feature scaling 'Year' and 'Population' columns
df['Year'] = (df['Year'] - df['Year'].min()) / (df['Year'].max() - df['Year'].min())
df['Population'] = (df['Population'] - df['Population'].min()) / (df['Population'].max() - df['Population'].min())
print(df)
```

```python
# Feature engineering - Creating population growth rate and average population features
df['Population_Growth_Rate'] = df.groupby('District')['Population'].pct_change()
df['Average_Population'] = df.groupby('District')['Population'].transform('mean')
print(df.head())
```

```python
# Removing missing values
df.fillna(0, inplace=True)
print(df.head())
```
